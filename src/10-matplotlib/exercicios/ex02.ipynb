```python
import pandas as pd 
import matplotlib.pyplot as plt

data = pd.read_csv('../../datasets/metrics.csv')
epochs = range(1, len(data) + 1)

plt.style.use('seaborn-v0_8')
```

Gere um gráfico semelhante ao de referência com as curvas de train_loss e val_loss
e train_acc e val_acc. Use o eixo x como épocas. Use o arquivo de dados
metrics.csv.

```python
fig, (ax2, ax1) = plt.subplots(1, 2, figsize=(16, 6))

ax1.plot(epochs, data['train_loss'], label='train', color='#66a8cc', linewidth=2.5)
ax1.plot(epochs, data['val_loss'], label='valid', color='#de7149', linewidth=2.5, linestyle='--')
ax1.set_title('model loss', fontsize=14, fontweight='bold')
ax1.set_xlabel('epoch', fontsize=12)
ax1.set_ylabel('loss', fontsize=12)
ax1.legend()
ax1.grid(True, linestyle='--', alpha=0.6)

ax2.plot(epochs, data['train_acc'], label='train', color='#66a8cc', linewidth=2.5)
ax2.plot(epochs, data['val_acc'], label='valid', color='#de7149', linewidth=2.5, linestyle='--')
ax2.set_title('model accuracy', fontsize=14, fontweight='bold')
ax2.set_xlabel('epoch', fontsize=12)
ax2.set_ylabel('accuracy', fontsize=12)
ax2.legend()
ax2.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
