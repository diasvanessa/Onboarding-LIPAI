1. Escolha um novo dataset e construa um programa que calcule as medidas estatísticas. Procure usar as funções apresentadas nessa atividade, assim como, outras que estão disponíveis nas bibliotecas do Python (https://www.scaler.com/topics/pandas/statistical-functions-in-pandas/).

```python
import pandas as pd
```

```python
data = pd.read_csv('../datasets/winequality-red.csv', sep = ';')
```

```python
data.head(5)
```

```python
data.columns = ['fixedAcidity', 'volatileAcidity', 'citricAcid', 'residualSugar', 'chlorides','freesulfurDioxide', 'totalsulfurDioxide', 'density', 'pH', 'sulphates', 'alcohol', 'quality']
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
from pandas.api.types import is_numeric_dtype
for column in data.columns:
    if is_numeric_dtype(data[column]):
        print('%s: ' % (column))
        print('\t Media = %.2f' % data[column].mean())
        print('\t Mediana = %.2f' % data[column].median())

        print('\t Desvio padrao = %.2f' % data[column].std())
        print('\t Variância = %.2f' % data[column].var())
        print('\t Minimo = %.2f' % data[column].min())
        print('\t Maximo = %.2f' % data[column].max())
```

2. Use o dataset escolhido e mostre por meio das representações gráficas (histogramas) as principais características. Procure usar as funções apresentadas nesta atividade.

3. O dataset Iris não possui valores ausentes. A base escolhida também pode conter todos os descritores por cada amostra. Use um dataset e sobre as amostras impute valores ausentes usando a média, mediana ou outras medidas.

```python
print(data.isnull().sum()) # o dataset nao possui valores nulos
# simulando se houvesse valores nulos:

media = data['fixedAcidity'].mean()
data['fixedAcidity'] = data['fixedAcidity'].fillna(media)

mediana = data['fixedAcidity'].median()
data['fixedAcidity'] = data['fixedAcidity'].fillna(mediana)
```

5. Compare os resultados obtidos usando um gráfico box plot. Que atributos possuem distribuições mais simétricas? Há a presença de valores extremos, denominados de outliers, para os atributos?

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Usar o estilo do Seaborn deixa o visual mais moderno
sns.set_theme(style="whitegrid")

fig, axes = plt.subplots(1, 3, figsize=(15, 6))

# Grupo com valores baixos
sns.boxplot(data=data[['volatileAcidity', 'citricAcid', 'chlorides', 'density', 'sulphates']], ax=axes[0], palette="tab10")
axes[0].set_title("Escala Baixa")

# Grupo com valores medios
sns.boxplot(data=data[['pH', 'alcohol', 'quality', 'fixedAcidity', 'residualSugar']], ax=axes[1], palette="Set2")
axes[1].set_title("Escala Média")

# Grupo com valores altos
sns.boxplot(data=data[['freesulfurDioxide', 'totalsulfurDioxide']], ax=axes[2], palette="Pastel1")
axes[2].set_title("Escala Alta")

for ax in axes:
    ax.tick_params(axis='x', rotation=45)

plt.show()
```

Os mais simétricos são os atributos 'pH', 'density' e 'quality'.
Há a presença de valores extremos em todos os atributos, mas a quantidade varia entre eles, sendo que os atributos 'chlorides' e 'residualSugar' possuem muitos valores extremos e 'alcohol' e 'quality' possuem poucos.
